# **Phase 1: Handling missing values**

```python
df.isnull().sum()
```

```python
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

# checking null values after handling missed values
df.isnull().sum()
```

# Phase 2: Handling duplicates

I have checked 'OrderID' column since it must be unique, in addition of checking the duplicate rows

```python
print("No. of duplicate rows:", df['OrderID'].nunique())
# duplicates in OrderID column
df.duplicated(subset=['OrderID']).sum()
```

```python
# repeated rows
df.duplicated().sum()
```

# **Phase 3: Speak One Language**

Checking date formats by checking 'Date' column datatype and if it is object, then there may be errors in formating, else, pandas has recognized that the data type of 'Date' is something with [ns]

```python
df['Date'].dtype
```

Ensuring that No invalid date formats are there in the dataset

```python
## this line isn't needed as we have already checked for Null values in the data
# print("No. of invalid dates:", df['Date'].isna().sum())

# Converting Date column format into ISO 8601
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

Stripping text columns

```python
# Removing any extra spaces in OrderID column
df['OrderID'] = df['OrderID'].astype(str).str.strip().str.upper()

# Removing any extra spaces in the object type columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()
```
